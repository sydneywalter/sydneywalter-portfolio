```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Assuming merged_data and label_encoders are already defined as per your code

# Initialize a new DataFrame to store the encoded data
encoded_data = merged_data.copy()

# Loop through each categorical column to encode
for column in encoded_data.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    encoded_data[column] = le.fit_transform(encoded_data[column])
    label_encoders[column] = le  # Store the label encoder for future reference



# Display the mapping of original labels to encoded labels
print("\nEncoding Mappings:")
encoding_mappings = pd.DataFrame(columns=['Feature', 'Encoded Label', 'Original Label'])
for feature, le in label_encoders.items():
    mapping_df = pd.DataFrame({'Feature': [feature] * len(le.classes_),
                               'Encoded Label': range(len(le.classes_)),
                               'Original Label': le.classes_})
    encoding_mappings = pd.concat([encoding_mappings, mapping_df], ignore_index=True)

print(encoding_mappings)

# Example usage of the get_encoded_value function
column_name = 'variant_2'
input_value = 'rs4680 AG'

try:
    encoded_value = label_encoders[column_name].transform([input_value])[0]
    print(f"\nEncoded value of '{input_value}' in column '{column_name}': {encoded_value}")
except KeyError:
    raise ValueError(f"Value '{input_value}' does not exist in the mapping for column '{column_name}'.")

# Print label_encoders to debug
print("\nLabel Encoders:")
for col, le in label_encoders.items():
    print(f"{col}: {le.classes_}")

```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Initialize LabelEncoders for each variant column and primary_antidepressant_choice
variant_columns = ['variant_1', 'variant_2', 'variant_3', 'variant_4', 'variant_5']
label_encoders = {}
for column in variant_columns + ['primary_antidepressant_choice']:
    label_encoders[column] = LabelEncoder()
    merged_data[column] = label_encoders[column].fit_transform(merged_data[column].astype(str))

# Display the encoded merged_data DataFrame
print("Encoded Merged Data:")
print(merged_data)

# Define X (features) and y (target) for merged_data
X = merged_data.drop(['patient_id', 'primary_antidepressant_choice'], axis=1)
y = merged_data['primary_antidepressant_choice']

# Initialize Logistic Regression model
logreg = LogisticRegression(random_state=42)
logreg.fit(X, y)

# Predict antidepressant choice for new_patient
# Ensure new_patient has all variant columns
new_patient = pd.DataFrame({
    'variant_1': [4],
    'variant_2': [10],
    'variant_3': [5],
    'variant_4': [7],
    'variant_5': [None],
    'primary_antidepressant_choice': [None]  # Placeholder for prediction
})

# Encode new_patient based on fitted LabelEncoders
for column in new_patient.columns:
    if column in label_encoders:
        new_patient[column].fillna('Unknown', inplace=True)
        # Handle previously unseen labels
        new_patient[column] = new_patient[column].apply(lambda x: label_encoders[column].transform([x])[0]
                                                        if x in label_encoders[column].classes_ else -1)

# Predict antidepressant choice
# Fill NaN with 'None' for new_patient for handling missing values
prediction = logreg.predict(new_patient[variant_columns].fillna('None'))

# Decode prediction back to original label
predicted_antidepressant = label_encoders['primary_antidepressant_choice'].inverse_transform(prediction)[0]

# Display prediction
print(f"Predicted Antidepressant Choice for the New Patient: {predicted_antidepressant}")

# Print accuracy and classification report (for demonstration purposes, should use test data for real evaluation)
accuracy = accuracy_score(y, logreg.predict(X))
print(f"Accuracy on the training set: {accuracy:.2f}")

# Classification report
print("Classification Report:")
print(classification_report(y, logreg.predict(X), target_names=label_encoders['primary_antidepressant_choice'].classes_))
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
# Print accuracy and classification report (for demonstration purposes, should use test data for real evaluation)
accuracy = accuracy_score(y, logreg.predict(X))
print(f"Accuracy on the training set: {accuracy:.2f}")

# Classification report
print("Classification Report:")
print(classification_report(y, logreg.predict(X), target_names=label_encoders['primary_antidepressant_choice'].classes_))

# Compute the confusion matrix
conf_matrix = confusion_matrix(y, logreg.predict(X))

# Plot the confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoders['primary_antidepressant_choice'].classes_, yticklabels=label_encoders['primary_antidepressant_choice'].classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
